```python
clean_europe = clean_europe.drop(['Province/State'], axis=1)
country_df = clean_europe.set_index("Country/Region")
```

```python
country_df = country_df.drop(['Lat', 'Long'], axis=1)
country_df
```

```python
clean_europe_tp = country_df.T
clean_europe_tp.head(30)
```

```python
# Multiple row and column selections using iloc and DataFrame

weekly_df = clean_europe_tp.iloc[[2,9,16,23,30,37,44,51,58,65,72,79,86,93,100,107,114,121,128,135,142,149,156,163,170,177,184], :] # 1st, 4th, 7th, 25th row + 1st 6th 7th columns.
weekly_df
```

```python
weekly_df.plot();
plt.title("Covid 19 Confirmed Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Cases")
```

```python
percent_df = weekly_df.pct_change()
percent_df
#Russia_Percent = data_file_df["Russia"]/1000
#data_file_df["Thousands of Dollars"] = thousands_of_dollars
```

```python
percent_df = percent_df.dropna(how='any')
percent_df
```

```python
percent_df.plot();
plt.xlim(2, 20)
plt.ylim(0, 1)
plt.title("Covid 19 Confirmed Cases Percent Change")
plt.xlabel("Date")
plt.ylabel("Percentage")
```
